## 2. Subplot (vertical)

```
import matplotlib.pyplot as plt
import numpy as np
```

```
x = [1,2,3,4,5]
y1 = [6,7,8,9,10]
y2 = [6.5,7.5,8.5,9.5,10.5]
y3 = [7,8,9,10,11]
```

```
fig, axs = plt.subplots(3)
fig.suptitle('Vertically stacked subplots')
axs[0].plot(x, y1)
axs[1].bar(x, y2)
axs[2].scatter(x, y3)
```

## 3. Subplot (vertical & horizontal)

```
import matplotlib.pyplot as plt
import numpy as np
```

```

x = [1,2,3,4,5]
y1 = [6,7,8,9,10]
y2 = [6.5,7.5,8.5,9.5,10.5]
y3 = [7,8,9,10,11]
```

```
fig, ax = plt.subplots(2,2)
fig.suptitle('Horizontally stacked subplots')
ax[0][0].plot(x, y1,color='red')
ax[0][1].bar(x, y2)
ax[1][0].scatter(x, y3,color='green')

ax[1][1].plot(x, y3)
```

# Workshop 1 : Application to Iris_Data

#### 1. Plot the bar that count the length of each species

```
# Import Libraries

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.preprocessing import LabelEncoder
```

```
# Prepare the Iris_Dataset

Iris_data = pd.read_csv("/content/drive/MyDrive/Data /iris.data",header = None)

Iris_data.columns = ['Sepal length','Sepal width','Petal length','Petal width','Species']
```

```
Iris_data.head()
```

```
# LabelEncoder to make Categories

label_encoder = LabelEncoder()
label = label_encoder.fit_transform(Iris_data['Species'])

x = label_encoder.classes_
```

```
class0 = len(Iris_data[Iris_data['Species'] == 'Iris-setosa'])
class1 = len(Iris_data[Iris_data['Species'] == 'Iris-versicolor'])
class2 = len(Iris_data[Iris_data['Species'] == 'Iris-virginica'])

y = [class0,class1,class2]
```

```
plt.bar(x,y,color=['red', 'blue', 'yellow'])
```
