```python
import pandas as pd

train_iris = pd.read_csv("/content/drive/MyDrive/Data /Iris/Train_Set.csv")
test_iris = pd.read_csv("/content/drive/MyDrive/Data /Iris/Test_Set.csv")

#test_iris
```

```python
test_iris
```

```python
print("Train set missing value: \n", train_iris.isnull().any())

print("Test set missing value: \n", test_iris.isnull().any())

# No missing value
```

```python
# preprocessing training set

train_y = train_iris.pop('Target')
train_X = train_iris
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=7)

model.fit(train_X,train_y)
```

```python
# The prediction of the data of test set of index 0 to 4

model.predict(test_iris.iloc[0:4,:-1])
```

```python
# Here's the accuracy of KNN with K=7

data_real = test_iris.iloc[:,-1]
data_pred = model.predict(test_iris.iloc[:,:-1])

from sklearn.metrics import accuracy_score

accuracy_score(data_real,data_pred)
```

```python
xy = {'x': [20,50,18,29,33,61,64,69,72,12],
      'y': [36,20,52,54,46,8,19,7,24,39]}

data_2 = pd.DataFrame(xy)
```

> 1st Iteration

```python
import numpy as np

first_iter = data_2.copy()


first_iter['C1_distance'] = np.nan
first_iter['C2_distance'] = np.nan
```

```python


# 1st iteration

# C1 = (10,50)

for i in range(len(data_2)):
  first_iter.iloc[i,-2] = np.linalg.norm([first_iter.iloc[i,0]-10, first_iter.iloc[i,1]-50])

# C2 = (50,20)

for i in range(len(data_2)):
   first_iter.iloc[i,-1] = np.linalg.norm([first_iter.iloc[i,0]-50, first_iter.iloc[i,1]-20])
```

```python
first_iter
```

```python
first_iter['group'] = np.nan
```

```python
for i in range(len(data_2)):
  if first_iter.iloc[i,-2] < first_iter.iloc[i,-3]:
    first_iter.iloc[i,-1] = 'B'

  else:
    first_iter.iloc[i,-1] = 'A'
    
```

```python
first_iter
```



---



> 2nd Iteration

> Now find the new C1 & C2 by finding the mean of group A & B

```python
# new C1

first_iter[first_iter.group == "A"].iloc[:,0:2].mean()
```

```python
# new C2

first_iter[first_iter.group == "B"].iloc[:,0:2].mean()
```

> C1 = (21.2,45.4) <br>
> C2 = (63.2,15.6)

```python
second_iter = data_2.copy()


second_iter['C1_distance'] = np.nan
second_iter['C2_distance'] = np.nan
```

```python
# 2nd iteration

# C1 = (63.2,15.6)

for i in range(len(data_2)):
  second_iter.iloc[i,-2] = np.linalg.norm([second_iter.iloc[i,0]-21.2, second_iter.iloc[i,1]-45.4])

# C2 = (21.2,45.4)

for i in range(len(data_2)):
   second_iter.iloc[i,-1] = np.linalg.norm([second_iter.iloc[i,0]-63.2, second_iter.iloc[i,1]-15.6])
```

```python
second_iter['group'] = np.nan
```

```python
for i in range(len(data_2)):
  if second_iter.iloc[i,-2] < second_iter.iloc[i,-3]:
    second_iter.iloc[i,-1] = 'B'

  else:
    second_iter.iloc[i,-1] = 'A'
    
```
