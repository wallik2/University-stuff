# 3)  Inheritance
---


> A way of creating new class for using details of existing class of existing class without modifying it

> Two types of class: 

1.  Base class (Parent class)
2.  Derived  Class (Child class)

image.png

```
#Parent Class

class Bird:

  def __init__(self):
    print("Bird is Ready!!")
  
  def whoisThis(self):
    print("Bird")
  
  def swim(self):
    print("Swim faster")
  
#Child class

class Penguin(Bird):                  #This is the key that Bird and Penguin are related -> Penguin(Bird)
  def __init__(self):
    super().__init__()            #Used to inherit __init__ contents from the parent class to this child class   
    print("Penguin is Ready!!")
  
  def whoisThis(self):
    super().whoisThis()          #unplug this and run peggy.whoisThis() to see the result
    print("Penguin")
  
  def run(self):
    print("Run faster")
```

```
peggy = Penguin()
```

```
peggy.whoisThis()
```

```
peggy.swim()  
```

```
peggy.run()
```

# 4) Encapsulation


---


> A way to prevent data from direct modification
> In the following case, we will lock the Maxprice cannot be directly changed

image (3).png
- As you can see the transparent side is public, and the red side is private which don't want the publics to change

- If __name__ = __main__ is also a method of encapsulation as the user import the module that have the method, but the user didn't run the module directly




```
class Computer:
  def __init__(self):
    self.__maxprice = 900            #maxprice as PRIVATE attributes
  
  def sell(self):
    print("Selling Price: {}".format(self.__maxprice))
  
  def setMaxPrice(self,price):                            #Setter Function: protect your data, particularly when creating classes. For each instance variable, a getter method returns its value while a setter method sets or updates its value. Getters and setters allow control over the values
    self.__maxprice = price
  
```

```
c = Computer()

c.sell()
```

```
c.__maxprice = 1000

c.sell()
```

```
c.setMaxPrice(1000)

c.sell()
```

```
c.sell()        #Now the maxprice is permanently changed to 1000

# : Encapsulation  Model    

```

# 5) Polymorphism


---

> An ability to use common interface for multiple form

---


image (2).png




```
class Parrot:
  
  def fly(self):
    print("Parrot can fly")
  
  def swim(self):
    print("Parrot can't swim ")


class Penguin:  
    
    def fly(self):
      print("Penguin can't fly")
    def swim(self):
      print("Penguin can swim")

#common interface
def flying_test(bird):
  bird.fly()

blu = Parrot()
peggy = Penguin()
```

```
#Passing the object
flying_test(blu)
flying_test(peggy)
```

#### Analysis: As you can see Parrot & Penguin each have common method *fly()* method. However, their functions are different. To allow polymorphism, we created common interface; flying_test() function that can take any object. Then we passed the object blu and peggy in the flying_test(). It ran effectively
